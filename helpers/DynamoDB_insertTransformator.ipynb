```python
def docts_converter(lable, days):
    data = dict()
    result = []
    lable.update(days)
    list = lable.keys()
    for key in list:
        data.setdefault(key, dict()).setdefault("label", dict())['S'] = key
        data[key].setdefault("language", dict()).setdefault('M', dict())
        data_M = data[key]["language"]['M']
        for k, v in lable[key].items():
            data_M.setdefault(k, dict(S=v))
    result.append(data)
    return result
    


```
